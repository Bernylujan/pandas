```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("DatosEstand.csv")
data
```

```python
plt.scatter(data["tasaInteres"],data["volume"])
plt.xlabel("Tasa de Interes")
plt.ylabel("Volumen de Operaciones")
plt.show
```

```python
plt.scatter(data["tasaInteres"],data["close"])
plt.xlabel("Tasa de Interes")
plt.ylabel("Cierre de Cotizaciones")
plt.show
```

```python
plt.scatter(data["tasaInteres"],data["adjClose"])
plt.xlabel("Tasa de Interes")
plt.ylabel("Cierre Ajustado de Cotizaciones")
plt.show
```
