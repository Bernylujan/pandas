# Graficas de datos investigados

{Explicacion de datos} 

#### Importamos las librerias necesarias 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

#### Cargamos nuestra tabala de datos, identificando la columna "fecha" como de tipo fecha

```python
data = pd.read_csv("DatosCSV.csv", parse_dates = ["fecha"])
data
```

#### Cargamos nuestra tabla pero con la columna fecha como index de la grafica (eje de las x)

```python
data = data.set_index("fecha")
data
```

#### Grafico de Tasas de Interes 

```python
plt.plot(data.tasaInteres)

plt.xlabel(" Fecha ") 
plt.ylabel( "Tasa deInteres" ) 
plt.show()
```

#### Grafica de Volumen de Operaciones de IPC

```python
plt.plot(data.volume)

plt.xlabel(" Fecha ") 
plt.ylabel( "Volumen de Operaciones de IPC" ) 
plt.show()
```

####  Grafica de Cierre de Cotizaciones de IPC

```python
plt.plot(data.close)

plt.xlabel(" Fecha ") 
plt.ylabel( "Cierre de Cotizaciones de IPC" ) 
plt.show()
```

####  Grafica de Cierre Ajustado de Cotizaciones de IPC

```python
plt.plot(data.adjClose)

plt.xlabel(" Fecha ") 
plt.ylabel( "Cierre Ajustado de Cotizaciones de IPC" ) 
plt.show()
```
