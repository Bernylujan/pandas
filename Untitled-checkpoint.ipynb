```python
miDataset = pd.read_csv("DatosCSV.csv", parse_dates = ["fecha"])
miDataset
```

```python
miDataset.info()
```

```python
plt.plot(miDataset.tasaInteres) 

plt.xlabel("Index") 
plt.ylabel("Tasa de Interes") 
plt.show
```

```python
miDataset = miDataset.set_index("fecha")
miDataset
```

```python
plt.plot(miDataset.tasaInteres)
plt.xlabel(" Fecha ") 
plt.ylabel(" Tasa de Interes ") 
plt.show()
```

```python
plt.plot(miDataset.volume)
plt.xlabel(" Fecha ") 
plt.ylabel(" Volumen de Operaciones IPC ") 
plt.show()
```

```python
plt.plot(miDataset.close)
plt.xlabel(" Fecha ") 
plt.ylabel(" Cierre de Cotizaciones IPC ") 
plt.show()
```

```python
plt.plot(miDataset.adjClose)
plt.xlabel(" Fecha ") 
plt.ylabel(" Cierre Ajustado de Cotizaciones IPC ") 
plt.show()
```
